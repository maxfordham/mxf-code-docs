````python
#| echo: false

import subprocess
import pandas as pd
from io import StringIO
from IPython.display import display, Markdown
import itables

def read_tsv_string(text, columns=None, **kwargs):
    """
    Parse a tab-separated string into a pandas DataFrame.

    Args:
        text: Tab-separated string
        columns: Optional list of column names
        **kwargs: Additional arguments for pd.read_csv

    Returns:
        pandas DataFrame
    """
    df = pd.read_csv(
        StringIO(text.strip()),
        sep='\t',
        header=None if columns else 0,
        names=columns,
        **kwargs
    )
    return df 

def display_df(df):
    df['last_updated'] = pd.to_datetime(df['last_updated']).dt.date
    df['repository'] = df['repository'].apply(lambda x: f'<a href="https://github.com/{x}" target=_blank>{x}</a>')
    return itables.show(df, 
             paging=False, 
             pageLength=-1, 
             style="text-align: left;", 
             classes="display compact", 
             columnDefs=[{"width": "100px", "targets": [2, 3]}, {"width": "300px", "targets": [0]}, {"className": "dt-left", "targets": "_all"}],
             allow_html=True
            )



def display_cmd(cmd):
    return display(Markdown(f"""```bash
{cmd}
```"""))

def display_gh_result(cmd):
    result = subprocess.run(cmd, capture_output=True, text=True, shell=True)
    df = read_tsv_string(
        result.stdout,
        columns=['repository', 'description', 'visibility', 'last_updated']
    )
    display_cmd(cmd)
    return display_df(df)
````
